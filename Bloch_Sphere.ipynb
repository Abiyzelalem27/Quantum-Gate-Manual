```python
import qutip as qt
import matplotlib.pyplot as plt

# 1. Define your specific state: |psi> = (1|0> - sqrt(3)|1>) / N
c0 = 1.0
c1 = -np.sqrt(3)
N = np.sqrt(abs(c0)**2 + abs(c1)**2) # N = 2
initial_state = (c0 * qt.basis(2, 0) + c1 * qt.basis(2, 1)) / N

# 2. Apply the Pauli Z operator
# This flips the phase (the 'lay' of the vector)

final_state = Z * initial_state

# 3. Visualization
b = qt.Bloch()

# We set a view that makes the X-axis clear (side view)
b.view = [-60, 30] 

# Add the states: 
# Point 1 (Initial): Should be at the back-bottom (Negative X)
# Point 2 (Final): Should be at the front-bottom (Positive X)
b.add_states(initial_state)
b.add_states(final_state)

b.show()
plt.show()

print(f"Normalization N is: {N}")
print(f"Initial State:\n{initial_state}")
print(f"Final State (after Z):\n{final_state}")
```
