```python
import numpy as np

# Basis
zero = np.array([1,0])
one = np.array([0,1])

# 3-qubit initial state |000>
initial_state = np.kron(np.kron(zero, zero), zero)

# Apply H on q0
H_q0 = np.kron(H, np.kron(I, I))
state_after_H = H_q0 @ initial_state

# CNOT q0 -> q2
CNOT_q0_q2 = np.kron(P0, np.kron(I, I)) + np.kron(P1, np.kron(I, X))
state_after_CNOT_q0_q2 = CNOT_q0_q2 @ state_after_H

# CNOT q0 -> q1
CNOT_q0_q1 = np.kron(P0, np.kron(I, I)) + np.kron(P1, np.kron(X, I))
state_after_q0_q1 = CNOT_q0_q1 @ state_after_CNOT_q0_q2

# CNOT q1 -> q0
CNOT_q1_q0 = np.kron(np.kron(I, P0), I) + np.kron(np.kron(I, P1), X)
state_after_CNOT_q1_q0 = CNOT_q1_q0 @ state_after_q0_q1

print(state_after_CNOT_q1_q0)
```

```python
import numpy as np
import qutip as qt
import matplotlib.pyplot as plt


# 1. Define your specific state
c0 = 1.0
c1 = -np.sqrt(3)
N = np.sqrt(abs(c0)**2 + abs(c1)**2)

initial_state = (c0 * qt.basis(2, 0) + c1 * qt.basis(2, 1)) / N

# Apply Z
final_state = Z * initial_state

# Visualization
b = qt.Bloch()
b.view = [-60, 30]

b.add_states(initial_state)
b.add_states(final_state)

b.show()
plt.show()

print("Normalization N is:", N)
print("Initial State:", initial_state)
print("Final State (after Z):", final_state)
```
