The POVM element Fi is associated with the measurement outcome i, such that the probability of obtaining it when making a measurement on the quantum state ρ is given by:
p(i)=tr(ρFi)

where tr is the trace operator. This is the POVM version of the Born rule. When the quantum state being measured is a pure state |ψ⟩  this formula reduces to:


p(i)=tr(|ψ⟩⟨ψ|Fi)= ⟨ψ|Fi|ψ⟩

## 4) Measurement function (Outcome + post-measurement state)

Goal: make a full measurement function that returns:
- outcome (k)
- probabilities p
- post-measurement state or  ρ_post

Steps:
1) Compute probabilities: pₖ = Tr(Pₖ ρ)
2) Sample the outcome k from p
3) Update the state using projection:

ρ_post = (Pₖ ρ Pₖ) / Tr(Pₖ ρ)

Return:
(outcome, ρ_post, p)


```python
F0 = np.array([[0.7, 0],
               [0,   0.2]], dtype=complex)

F1 = np.eye(2, dtype=complex) - F0   # because F0 + F1 = I

psi = np.array([1, 1], dtype=complex) / np.sqrt(2)   # (|0>+|1>)/√2
rho = np.outer(psi, psi.conj())                      # rho = |psi><psi|

p0 = np.trace(rho @ F0)
p1 = np.trace(rho @ F1)

print("p0 =", np.real(p0))
print("p1 =", np.real(p1))
```
