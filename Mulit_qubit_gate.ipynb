```python
import sys
import os

# Python path so notebook can find bell_state
sys.path.insert(0, os.path.abspath("src"))
                
import numpy as np  # arrays, linear algebra, numerical computing
import quantum_gate as qg #custom package(module)
```

```python
CZ=qg.operators.controlled_gate(qg.operators.Z, 0, 1, 2)
print("CZ=\n", CZ)
```

```python
S01 = qg.operators.P0 @ qg.operators.X @ qg.operators.P1   
S10 = qg.operators.P1 @ qg.operators.X @ qg.operators.P0   
SWAP = (np.kron(qg.operators.P0, qg.operators.P0) +
        np.kron(qg.operators.P1, qg.operators.P1) +
        np.kron(S01, S10) +
        np.kron(S10, S01))
print("SWAP =\n", SWAP)
```

```python
CCX=np.kron(qg.operators.P0,np.kron(qg.operators.I,qg.operators.I))+np.kron(qg.operators.P1,np.kron(qg.operators.P0,qg.operators.I))+np.kron(qg.operators.P1,np.kron(qg.operators.P1,qg.operators.X))
print("CCX =\n", CCX)
```

```python
initial_state = np.array([0,0,1,0,
                          0,0,0,0])

print("initial_state =", initial_state)

# Hadamard gate acting on qubit q0
H_q0 = np.kron(qg.operators.H,
               np.kron(qg.operators.I, qg.operators.I))

print("H_q0 =\n", H_q0)

state_after_H = H_q0 @ initial_state

print("state_after_H =", state_after_H)

# CNOT(q0 -> q2)
CNOT_q0_q2 = np.kron(qg.operators.P0,
                     np.kron(qg.operators.I, qg.operators.I)) + \
             np.kron(qg.operators.P1,
                     np.kron(qg.operators.I, qg.operators.X))

print("CNOT_q0_q2 =\n", CNOT_q0_q2)

state_after_CNOT_q0_q2 = CNOT_q0_q2 @ state_after_H

print("state_after_CNOT_q0_q2 =", state_after_CNOT_q0_q2)

# CNOT(q0 -> q1)
CNOT_q0_q1 = np.kron(qg.operators.P0,
                     np.kron(qg.operators.I, qg.operators.I)) + \
             np.kron(qg.operators.P1,
                     np.kron(qg.operators.X, qg.operators.I))

print("CNOT_q0_q1 =\n", CNOT_q0_q1)

state_after_q0_q1 = CNOT_q0_q1 @ state_after_CNOT_q0_q2

print("state_after_q0_q1 =", state_after_q0_q1)

# CNOT(q1 -> q0)
CNOT_q1_q0 = np.kron(np.kron(qg.operators.I, qg.operators.P0),
                     qg.operators.I) + \
             np.kron(np.kron(qg.operators.I, qg.operators.P1),
                     qg.operators.X)

print("CNOT_q1_q0 =\n", CNOT_q1_q0)

final_state = CNOT_q1_q0 @ state_after_q0_q1

print("final_state =", final_state)
```
